````python
def plot_decision_boundary(model, X, y):
  """
  Plots the decision boundary created by a model predicting on X.
   
  """
  # Define the axis boundaries of the plot and create a meshgrid
  
  x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
  y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min, y_max, 100))
  
  # Create X values (we're going to predict on all of these)
  x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2D arrays together: https://numpy.org/devdocs/reference/generated/numpy.c_.html
  
  # Make predictions using the trained model
  y_pred = model.predict(x_in)

  # Check for multi-class
  if len(y_pred[0]) > 1:
    print("doing multiclass classification...")
    # We have to reshape our predictions to get them ready for plotting
    y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
  else:
    print("doing binary classifcation...")
    y_pred = np.round(y_pred).reshape(xx.shape)
  
  # Plot decision boundary
  plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
````

## **Find best learning rate**


How to find best learning rate 

---


To find the Learning rate @ which the loss decreses a most
* callback :- we can create callbacks to find best learning rate



````python
tf.random.set_seed(42) # random seed

# create a model
model_6=tf.keras.Sequential([
  tf.keras.layers.Dense(10,activation='relu'),
  tf.keras.layers.Dense(1,activation='sigmoid')                             
])

# compile a model
model_6.compile(loss='binary_crossentropy',
                optimizer=tf.keras.optimizers.Adam(lr=0.01),
                metrics=['accuracy'])
# learning callback
lr_scheduler=tf.keras.callbacks.LearningRateScheduler(lambda epoch : 1e-4 * 10**(epoch/20))

# fit into model 
history_6=model_6.fit(x_train,y_train,epochs=100,callbacks=[lr_scheduler])

````

````python
 pd.DataFrame(history_6.history)
````

````python
pd.DataFrame(history_6.history).plot()
````

 This graph show that intially our loss decreases but later on it increases..

````python
lrs=1e-4 * 10**(tf.range(100)/20)
lrs
````

````python
# let 's plot it
plt.figure(figsize=(8,6))
plt.semilogx(lrs,history_6.history["loss"])
plt.xlabel("learning rate")
plt.ylabel("loss")
plt.title("Learning rate v/s Loss")
plt.show()
````

 we can see that our optimal learning rate in b/w `0.01-0.1`....

## **Building multi-class classification model**

````python
tf.random.set_seed(42)

model_8=tf.keras.Sequential([
  tf.keras.layers.Flatten(input_shape=(28,28)),                           
  tf.keras.layers.Dense(100,activation='relu'),
  tf.keras.layers.Dense(10,activation='softmax')                             
])

model_8.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=['accuracy'])

history_8=model_8.fit(train_data,train_labels,epochs=10)
````

**here we used SparseCategoricalCrossEntropy as loss function because our labels are not one_hot encoded**

**CategoricalCrossEntropy** requires labels in one_hot encoded form. 
let's find a way for it... 

````python
tf.random.set_seed(42)

model_9=tf.keras.Sequential([
  tf.keras.layers.Flatten(input_shape=(28,28)),
  tf.keras.layers.Dense(100,activation='relu'),
  tf.keras.layers.Dense(10,activation='softmax')                             
])

model_9.compile(loss=tf.keras.losses.CategoricalCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=['accuracy'])

history_9=model_9.fit(train_data,tf.one_hot(train_labels,depth=10),epochs=10)
````

````python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.utils.multiclass import unique_labels


def plot_confusion_matrix(y_test,y_pred,labels=None):
  display_labels=None
  
  confusion_matrix=tf.math.confusion_matrix(y_test,y_pred).numpy()
  cn_normalise=confusion_matrix.astype('float')/confusion_matrix.sum(axis=1)[:np.newaxis]

  if display_labels is None:
      
    if labels is None:
        display_labels = unique_labels(y_test, y_pred)
    else:
        display_labels = labels


  temp=display_labels.tolist()
  display=ConfusionMatrixDisplay(cn_normalise,display_labels=temp)
  
  display.plot(cmap=plt.cm.Blues_r)   
     
````

````python
plt.figure(figsize=(10,10))
plot_confusion_matrix(test_labels,y_preds,labels=np.array(items.values()))
````

````python
import matplotlib.pyplot as plt

def plot_random_image(model,images,labels,normal):
  
  plt.figure(figsize=(2,2))

  idx=np.random.randint(len(images))

  predicted=items[model.predict(normal)[idx].argmax()]
  confidence=tf.math.round(model.predict(normal)[idx].max()*100).numpy()
  if predicted==items[labels[idx]]:
    color="g"
  else:
    color='r'  
  plt.imshow(images[idx])
  plt.xlabel("predicted ="+predicted+"\nactual="+items[labels[idx]]+"\nconfidence="+str(confidence)+"%",c=color)
````
