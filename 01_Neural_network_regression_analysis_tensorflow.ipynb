## **Improving our model**

We can improve our model by altering the steps involved to create a model.

**Changes going to made :**
1. Creating a model üêæ Add more number of nuerons or hidden layers and change activation function of each layer.
2. Comiling a model üêæ here we can change optimization function and it's learning rate.
3. fitting a model üêæ We can increase the size of data or time of training.

**increasing time of training ; means epochs**

`````python
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1))

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

model.fit(X,Y,epochs=1000)
`````

`````python
model.predict([20]) # model's prediction
`````

‚úå here we get predicted value more close to actual value by increasing time of learning but it's too long... ‚è≤
1000 epochs

let's go one step downwards for time feasible... üîΩ epoch=100.

`````python
tf.random.set_seed(42)

model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1))

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

model.fit(X,Y,epochs=100)
`````

`````python
model.predict([20])
`````

we get good respone near to our actual value ... üòê

**now let's make another change ... add a hidden layer...** ‚ûï

`````python
tf.random.set_seed(42)

model=tf.keras.Sequential()

model.add(tf.keras.layers.Dense(100,activation='relu'))
model.add(tf.keras.layers.Dense(1))

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

model.fit(X,Y,epochs=100)
print("\n\n")
model.predict([20])
`````

Is it overfitting ‚ùì

`````python
tf.random.set_seed(42)

model=tf.keras.Sequential()

model.add(tf.keras.layers.Dense(100,activation=None))
model.add(tf.keras.layers.Dense(1))

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

model.fit(X,Y,epochs=100)
print('\n\n')
model.predict([20])
`````

üîë thing to Note here is earlier we use `relu` as activation function and now we didn't take any activation function.

but difference is appoaximately same , `Y(pred)-Y(actual)`.
image.png

'relu' üêæ But the issue is that all the negative values become zero immediately which decreases the ability of the model to fit or train from the data properly. That means any negative input given to the ReLU activation function turns the value into zero immediately in the graph, which in turns affects the resulting graph by not mapping the negative values appropriately.


**Change the optimizer**

`````python
tf.random.set_seed(42)

model=tf.keras.Sequential()

model.add(tf.keras.layers.Dense(100))
model.add(tf.keras.layers.Dense(1))

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['mae'])

model.fit(X,Y,epochs=100)

print('\n\n')

model.predict([20])
`````

**change the learning rate of optimizer**

`````python
tf.random.set_seed(42)

model=tf.keras.Sequential()

model.add(tf.keras.layers.Dense(100))
model.add(tf.keras.layers.Dense(1))

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(lr=0.01), # default=0.001 10x
              metrics=['mae'])

model.fit(X,Y,epochs=100)

print('\n\n')

model.predict([20])
`````

ü§© here we got so close to actual value we get our best model üéâ üéâ..

* make a function to plot 

`````python
def plot_predictions(train_data,train_labels,test_data,test_labels,prediction):
  """
    plot training,testing data and predictions...
  """

  plt.figure(figsize=(12,8))

  plt.scatter(train_data,train_labels,c='b',label='Training Data')
  plt.scatter(test_data,test_labels,c='g',label="Testing Data")
  plt.scatter(test_data,prediction,c='r',label="Predictions")

  plt.legend()
`````

`````python
# importing all reqi=uired libraries

import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
`````

* convert non-numeric to numeric by categorical to one hot encoded...

`````python
df=pd.get_dummies(insaurance) # one hot encoding
df.head()
`````

`````python
# Now create features and labels

x=df.drop("charges",axis=1)
y=df["charges"]
`````

`````python
# Now create training and testing dataset

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
`````

`````python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
`````

`````python
x_train=tf.constant(x_train)
y_train=tf.constant(y_train)
x_test=tf.constant(x_test)
y_test=tf.constant(y_test)
`````

`````python
x_train.ndim
`````

`````python
y_train.ndim
`````

`````python
# Now create a baseline model

model=tf.keras.Sequential([
   tf.keras.layers.Dense(10)
])

model.add(tf.keras.layers.Dense(1))

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

model.fit(x_train,y_train,epochs=100)
`````

üòû not good value for Mean Absolute error ... Now we have to improve it...

`````python
# Now create a baseline model

model=tf.keras.Sequential([
   tf.keras.layers.Dense(100,activation='relu')
])
model.add(tf.keras.layers.Dense(100,activation='relu'))
model.add(tf.keras.layers.Dense(1,activation='relu'))

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(lr=0.01),
              metrics=['mae'])

history=model.fit(x_train,y_train,epochs=500)
`````

`````python
model.evaluate(x_test,y_test)
`````

## **preprocessing of data**

1. **Normalisation** üêæ  The goal of normalization is to change the values of numeric columns in the dataset to use a common scale, without distorting differences in the ranges of values or losing information.
2. **Standardisation** üêæ  Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation. 

`````python
x
`````

`````python
x["age"].plot(kind='hist')
`````

`````python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split

# transform our data
ct=make_column_transformer(
    (MinMaxScaler(),['age','bmi','children']), # normalise it in range 0-1
    (OneHotEncoder(handle_unknown='ignore'),['sex','smoker','region'])
)

# Create X & y
x=insaurance.drop("charges",axis=1)
y=insaurance["charges"]

# split the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

ct.fit(x_train)

x_train_normalise=ct.transform(x_train)
x_test_normalise=ct.transform(x_test)
`````

`````python
x_train.shape,x_train_normalise.shape
`````

* now build a neural network for this normalised data

`````python
model_nmsd=tf.keras.Sequential([])

model_nmsd.add(tf.keras.layers.Dense(100,activation='relu'))
model_nmsd.add(tf.keras.layers.Dense(10,activation='relu'))
model_nmsd.add(tf.keras.layers.Dense(1,activation='relu'))

model_nmsd.compile(loss=tf.keras.losses.mae,
                   optimizer=tf.keras.optimizers.Adam(lr=0.01),
                   metrics=['mae'])

model_nmsd.fit(x_train_normalise,y_train,epochs=500)
`````

`````python
model_nmsd.evaluate(x_test_normalise,y_test)
`````
